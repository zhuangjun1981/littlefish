```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
```

```python
marker_size = 1000
color_mother = "#CD5C5C"
color_ko = "#999999"
color_child = "#4169E1"
color_lifespan = "darkorchid"
```

```python
f, ax = plt.subplots(figsize=(14, 6))

ax.scatter(range(4), [2] * 4, marker='s', s=marker_size, fc=color_mother, ec=color_mother)
ax.scatter(range(4, 10), [2] * 6, marker='s', s=marker_size, fc=color_ko, ec=color_ko)

ax.scatter(range(4), [0.5] * 4, marker='s', s=marker_size, fc=color_mother, ec=color_mother)
ax.scatter(range(4, 10), [0.5] * 6, marker='s', s=marker_size, fc=color_child, ec=color_child)

ax.plot([-0.3, 3.3], [1.8, 1.8], ls='-', color=color_mother, lw=3)
ax.plot([3.7, 9.3], [1.8, 1.8], ls='-', color=color_ko, lw=3)

ax.plot([-0.3, 3.3], [0.3, 0.3], ls='-', color=color_mother, lw=3)
ax.plot([3.7, 9.3], [0.3, 0.3], ls='-', color=color_child, lw=3)


arrow_lifespan = FancyArrowPatch(
    (8.5, 2.4), (0.5, 2.4), arrowstyle='<->', mutation_scale=25,
    color=color_lifespan, lw=4
)
ax.add_patch(arrow_lifespan)

arrow_copy = FancyArrowPatch(
    (1.5, 1.5), (1.5, 0.8), arrowstyle='->', mutation_scale=25,
    color=color_mother, lw=4
)
ax.add_patch(arrow_copy)
ax.text(1.4, 1.2, "copy", fontsize=14, ha='right', va='center', weight='semibold', color=color_mother)

arrow_spawn = FancyArrowPatch(
    (3, 1.5), (6.5, 0.8), arrowstyle='->', mutation_scale=25,
    color=color_child, lw=4
)
ax.add_patch(arrow_spawn)
ax.text(5, 1.2, "spawn with mutation", fontsize=14, ha='left', va='center', weight='semibold', color=color_child)


ax.text(-0.3, 2.4, "Long", fontsize=14, ha='left', va='center', weight='semibold', color=color_lifespan)
ax.text(9.3, 2.4, "Short", fontsize=14, ha='right', va='center', weight='semibold', color=color_lifespan)
ax.text(4.5, 2.45, "Lifespan", fontsize=14, ha='center', va='bottom', weight='semibold', color=color_lifespan)

ax.text(1.5, 1.7, "Mother Fish", color=color_mother, fontsize=14, ha='center', va='center', weight='semibold')
ax.text(6.5, 1.7, "Fish that have no offspring", color=color_ko, fontsize=14, ha='center', va='center', weight='semibold')

ax.text(6.5, 0.2, "Child Fish", color=color_child, fontsize=14, ha='center', va='center', weight='semibold')

ax.text(-0.7, 2, "Generation n", color="#555555", fontsize=14, ha='right', va='center', weight='semibold')
ax.text(-0.7, 0.5, "Generation n+1", color="#555555", fontsize=14, ha='right', va='center', weight='semibold')

ax.set_ylim(0, 2.6)
ax.set_xlim(-3, 9.5)
ax.set_axis_off()
```
