```python
import littlefish.core.fish as fi
import littlefish.core.plotting as pl
import littlefish.core.evolution as evo
import random
import numpy as np

random.seed(0)
np.random.seed(0)
```

```python
fish = fi.generate_standard_fish()
brain_mutation = evo.get_default_brain_mutation()
fish = evo.mutate_fish(
    fish, 
    brain_mutation=brain_mutation, 
    neuron_mutation_rate=1.,
    connection_mutation_rate=1., 
    verbose=True
)
```

```python
brain = fish.get_brain()
pl.plot_brain(
    brain,
    neuron_label_kws = {
        'fontsize': 15, 'va': 'center', 'ha': 'center', 
        'color': '#000000', 'family': 'monospace', 'weight': 'semibold'
    },
    neuron_kws = {'mew': 3, 'ms': 80.},
)
```
